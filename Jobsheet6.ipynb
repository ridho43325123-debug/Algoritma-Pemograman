Pratikum 05

```python
A = [[1, 2, 3], [4, 5, 6]]
B = [[7, 8, 9], [10, 11, 12]]
C = []

for i in range(len(A)):
    baris = []
    for j in range(len(A[0])):
        baris.append(A[i][j] + B[i][j])
    C.append(baris)

print(C)
```

Pratikum 06

```python
nilai = [
     [80, 90, 85],
     [70, 75, 80],
     [88, 92, 84]
]

for i, baris in enumerate(nilai, start=1):
     rata = sum(baris) / len(baris)
     print(f"Rata-rata mahasiswa {i}: {rata:.2f}")

```

Penugasan 05

```python
A = [[1, 2],
     [3, 4]]

B = [[5, 6],
     [7, 8]]

C = []

for i in range(len(A)):
    baris = []
    for j in range(len(A[0])):
        baris.append(A[i][j] + B[i][j])
    C.append(baris)

print("Hasil penjumlahan matriks:")
print(C)
```

Penugasan 06

```python
nilai = [
    [75, 80, 85],
    [60, 70, 65],
    [90, 95, 92]
]

for i, mhs in enumerate(nilai, start=1):
    rata = sum(mhs) / len(mhs)
    print(f"Mahasiswa {i} - Rata-rata: {rata:.2f}")
```

Penugasan 07

```python
stok = [["Buku", 20], ["Pulpen", 35], ["Penghapus", 15]]

print("Daftar Stok Barang:")
for barang, jumlah in stok:
    print(f"{barang}: {jumlah}")

# Cari stok terbanyak
barang_terbanyak = max(stok, key=lambda x: x[1])
print(f"\nBarang dengan stok terbanyak: {barang_terbanyak[0]} ({barang_terbanyak[1]} unit)")
```
