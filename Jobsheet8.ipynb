
Praktikum 07 Fungsi Decorator


```python
def decorator_sapa(fungsi):
    def wrapper():
        print("Sebelum fungsi dijalankan...")
        fungsi()
        print("Setelah fungsi dijalankan...")
    return wrapper

@decorator_sapa
def sapa():
    print("Halo, Mahasiswa Polines!")

sapa()
```

Penugasan 8

```python
def logger(func):
    def wrapper():
        print("Memulai fungsi...")
        result = func()
        print("Fungsi selesai dijalankan.")
        return result
    return wrapper

@logger
def halo():
    print("Halo! Ini sapaan sederhana.")

halo()
```

Penugasan 9

```python
# a. Fungsi input nilai
def input_nilai():
    nilai = []
    banyak = int(input("Masukkan jumlah nilai: "))
    for i in range(banyak):
        n = float(input(f"Nilai ke-{i+1}: "))
        nilai.append(n)
    return nilai

# b. Fungsi hitung rata-rata
def hitung_rata(nilai):
    return sum(nilai) / len(nilai)

# c. Fungsi status kelulusan
def status_kelulusan(rata):
    if rata >= 60:
        return "Lulus"
    else:
        return "Tidak Lulus"

# d. Lambda nilai akhir (40% tugas, 60% ujian)
nilai_akhir = lambda tugas, ujian: (0.4 * tugas) + (0.6 * ujian)

# --- Contoh penggunaan (tanpa input biar langsung jalan) ---
nilai_mahasiswa = [80, 70, 60]
rata = hitung_rata(nilai_mahasiswa)

print("Nilai:", nilai_mahasiswa)
print("Rata-rata:", rata)
print("Status:", status_kelulusan(rata))
print("Nilai Akhir:", nilai_akhir(75, 85))
```
