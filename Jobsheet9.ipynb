Penugasan 04

```python
%%writefile konversi.py
def celcius_ke_fahrenheit(c):
    return (c * 9/5) + 32

def fahrenheit_ke_celcius(f):
    return (f - 32) * 5/9
```

```python
%%writefile main.py
import konversi

print("30°C =", konversi.celcius_ke_fahrenheit(30), "F")
print("100°F =", konversi.fahrenheit_ke_celcius(100), "C")
```

Penugasan 06

```python
%%writefile laporan.py
def tampilkan_laporan(jumlah_soal, benar):
    print("\n===== LAPORAN KUIS =====")
    print("Jumlah Soal:", jumlah_soal)
    print("Jawaban Benar:", benar)
    print("Skor:", (benar / jumlah_soal) * 100, "%")
```

```python
import random
import laporan

jumlah_soal = 5
benar = 0

for i in range(jumlah_soal):
    a = random.randint(1, 10)
    b = random.randint(1, 10)
    jawaban = int(input(f"Soal {i+1}: {a} + {b} = "))

    if jawaban == a + b:
        print("Benar!")
        benar += 1
    else:
        print("Salah! Jawaban yg benar adalah", a + b)

laporan.tampilkan_laporan(jumlah_soal, benar)
```
